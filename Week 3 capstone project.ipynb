<b>Create a URL<b>

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

<b>Import Libraries for requests and beautiful soup<b>

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

<b>Get a response from the wiki page<b>

```python
response = requests.get(url).text
```

<b>Check the response<b>

```python
response
```

<b>Clean the response by reading the html and loading it as csv to a dataframe<b>

```python
import pandas as pd

#Get the html source
html = response.encode("UTF-8")
df = pd.read_html(html)[0]
df.to_csv('beautifulsoup_pandas.csv',header=0,index=False)
df
```

<b>Create a neighborhood dataframe with the required columns<b>

```python
column_names=['Postal_code','Borough','Neighbourhood']
neighborhood = pd.DataFrame(columns=column_names)
```

```python

neighborhood
```

<b>Append all the columns in the dataframe to a single columns and in the next steps we will split that single column in to the required columns.<b>

```python
newn=[]
for i in range(9):
    for data in df[i]:
        newn.append(data)
```

```python
newn1=pd.DataFrame(newn)
```

```python
newn1
```

<b>Add necessary columns<b>

```python
newn2=newn1[0].str.split("(", n = 1, expand = True) 
newn1["PostalCode"]= newn2[0].str[:3]
newn1["Borough"]= newn2[0].str[3:]
newn1["Neighbourhood"]= newn2[1].replace('',')')
newn1
```

<b> Insert in to the neighborhood dataframe<b>

```python
neighborhood=newn1[['PostalCode','Borough','Neighbourhood']]
```

```python
neighborhood
```

<b> Clean up data. Delete records that have ['Borough']== 'Not assigned'. Remove sprcial characters and replace / with comma for neighborhoods<b>

```python
neighborhood.drop(neighborhood[neighborhood['Borough']== 'Not assigned'].index, inplace = True) 
neighborhood['Neighbourhood'] = neighborhood['Neighbourhood'].str.replace('/',',')
neighborhood['Neighbourhood'] = neighborhood['Neighbourhood'].str.replace(')','')
```

<b> Replace null values in Neighbourhood with Borough values<b>

```python

neighborhood['Neighbourhood'].fillna(neighborhood['Borough'], inplace = True) 
```

<b>Check for nulls<b>

```python
bool_series = pd.isnull(neighborhood['Neighbourhood']) 
  
# displayed data only with  NaN 
neighborhood[bool_series] 
```
